# Using Git with minimal password hassles from the DataHub

If you intend to work with git repositories from the Hub and perform operations that require authentication, such as cloning private repos or pushing to git, you will need to manage that authentication process from inside the hub.

For occasional use, you can get an [access token](https://docs.github.com/en/enterprise-server@3.4/authentication/keeping-your-account-and-data-secure/creating-a-personal-access-token), and use it as a password when required.

But for more regular usage, our DataHub has a very convenient feature: an [authentication manager](https://github.com/jupyterhub/gh-scoped-creds), that lets you do two things:

* Define which repositories you want to grant access to.
* Open a "tunnel" between the hub and those repositories that stays open for a specified amount of time. During that time, you can push/pull from the hub to those repos without having to input extra passwords.

## Initial setup

In order to use the system on the Data 100 Hub, you must first configure your GitHub account to allow access to the app. Note that this step only needs to be completed once for the entire organization (unless you want to change the access scope).
